```python
import requests
import pandas as pd
import numpy
import time
```

```python
#Keys
API_Key='XXXXXX' # Replace API key of your Google Account

# Follow the below link to generate API key for Google Acc
#https://www.slickremix.com/docs/get-api-key-for-youtube/

Channel_id='UC5EQWvy59VeHPJz8mDALPxg'  # Micset youtube Channel ID
```

```python
#Making second API call to get video statistics
def getVideoStats(video_id):
    url_video_stats="https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&key="+API_Key+"&part=statistics&order=date"
    response_video_stats=requests.get(url_video_stats).json()
    like_count=response_video_stats['items'][0]['statistics']['likeCount']
    view_count=response_video_stats['items'][0]['statistics']['viewCount']
    favourite_count=response_video_stats['items'][0]['statistics']['favoriteCount']
    comment_count=response_video_stats['items'][0]['statistics']['commentCount']
    return like_count,view_count,favourite_count,comment_count
```

```python
# Getting videoId, Title and Upload date for all the video in the channel
def getVideoDetails(df):
    url="https://www.googleapis.com/youtube/v3/search?key="+API_Key+"&part=snippet&order=date&channel_id="+Channel_id+"&maxResults=10000"
    responses=requests.get(url).json()
    time.sleep(1) #give it a second before entering for loop
    
    for video in responses['items']:
      if(video['id']['kind']=='youtube#video'):
        video_id= video['id'].get('videoId')
        Title_latest=video['snippet']['title']
        title=str(Title_latest).replace("&#39;", "")
        upload_date=video['snippet']['publishedAt']
        upload_date_trim= upload_date.split("T")[0]
      
        #Implementing good software engineering skills
        #Trying this second API call as a function
        like_count, view_count, favourite_count, comment_count= getVideoStats(video_id)
        #Making an second API Call to get views, dislikes and Comment count
      
      
        #Save data in Pandas Dataframe
        df=df.append({"video_id":video_id,
           "video_Title":title,
           "upload_date":upload_date_trim,
           "like_count":like_count,
           "view_count":view_count,
           "favourite_count":favourite_count,
           "comment_count":comment_count},ignore_index=True)

    return df
```

```python
#Building a dataFrame

df2=pd.DataFrame(columns=["video_id","video_Title","upload_date","like_count","view_count","favourite_count","comment_count"])
df2 = getVideoDetails(df2)
```
